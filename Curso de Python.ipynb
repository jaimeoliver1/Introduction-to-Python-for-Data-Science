## Colecciones

Hemos visto los diferentes tipos básicos que hay. Estos tipos se pueden organizar en diferentes tipos de colecciones: 
- **Listas**: 
  La lista es un tipo de colección ordenada. Sería equivalente a lo que en otros lenguajes se conoce por arrays, o vectores. Las listas pueden contener cualquier tipo de dato: números, cadenas, booleanos, … y también listas.
  
  
- **Tuplas**: Propiedades muy parecidas a las listas, pero algunas diferencias: 
  - Las tuplas son inmutables, una vez declaradas no pueden cambiarse sus elementos. 
  - Esto hace que las tuplas sean más rapidas que las listas, y que puedan utilizarse como claves de diccionarios (util).
  
  
- **Diccionarios**: Los diccionarios, también llamados matrices asociativas, deben su nombre a que son colecciones que relacionan una clave y un valor.

    - Como clave se puede utilizar cualquier valor inmutable:números, cadenas, booleanos, tuplas 
    - Los diccionarios no tienen orden, se accede a sus elementos por la clave
    - Se implementan como tablas hash, lo que les dota de una gran eficiencia.

## Orientación a objetos

Una de las ventajas de Python es que es muy versátil, y que acepta tanto programación funcional como programación orientada a objetos. 

La programación orientada a objetos es un paradigma, en el cual se trabaja con los siguientes conceptos: 

- Un **objeto** es una entidad que agrupa un estado y una funcionalidad.

- El estado del objeto se define a través de variables llamadas **atributos**. 

- Cada objeto tiene una serie de funciones definidas sobre el mismo que se denominan **métodos**. 

Un ejemplo de objeto podría ser un coche, que tiene una serie de atributos (marca, numero de puertas, tipo de carburante) y una serie de métodos (arrancar, parar). 


- Una **clase** es una plantilla a partir de la cual se instancian objetos. 

Volviendo al ejemplo. Hay muchos coches en el mundo (objetos), pero tienen todos unos atributos y metodos en comun. 

Podemos definir la clase coche. Para ello lo primero que se debe hacer es iniciarla. 

```python
class coche():
    '''Abstraccion de los objetos coche'''
    def __init__(self, gasolina):
        self.gasolina = gasolina
        print("Tenemos ",gasolina," litros")
        
```

A continuación se pueden definir algunos métodos:

```python
class coche():
    '''Abstraccion de los objetos coche'''
    def __init__(self, gasolina):
        self.gasolina = gasolina
        print("Tenemos ",gasolina," litros")
    
    def arrancar(self):
        if self.gasolina >0:
            print('Arranca')
        else:
            print('No arranca')
            
    def conducir(self):
        if self.gasolina >0:
            self.gasolina -= 1
            print("Quedan ", self.gasolina," litros")
        else:
            print("Repostar")
            
    def repostar(self, repost):
        self.gasolina += repost
        print("Quedan ", self.gasolina," litros")
```

```python
coche_a = coche(2)
coche_a.arrancar()
coche_a.conducir()
coche_a.conducir()
coche_a.conducir()
coche_a.repostar(10)
```

Ejemplo de la cuenta bancaria.

```python
class cuenta:
    '''Objeto cuenta corriente'''
    def __init__(self,ingreso_inicial):
        self.saldo = ingreso_inicial
        
    def ingreso(self,cantidad):
        self.saldo += cantidad
        
    def transferencia(self,cantidad2):
        self.saldo -= cantidad2
        
    def consultar_saldo(self):
        print('Su saldo es de',self.saldo)
        
```

```python
cuenta_juan = cuenta(500)
cuenta_juan.consultar_saldo()

cuenta_juan.ingreso(10000)
cuenta_juan.consultar_saldo()
cuenta_juan.transferencia(100)
cuenta_juan.consultar_saldo()
```

```python
class vehiculo():
    '''Abstraccion de los objetos coche'''
    def __init__(self, gasolina):
        self.gasolina = gasolina
        print("Tenemos ",gasolina," litros")
    
    def arrancar(self):
        if self.gasolina >0:
            print('Arranca')
        else:
            print('No arranca')
            
    def conducir(self):
        if self.gasolina >0:
            self.gasolina -= 1
            print("Quedan ", self.gasolina," litros")
        else:
            print("Repostar")
            
    def repostar(self, repost):
        self.gasolina += repost
        print("Quedan ", self.gasolina," litros")
        
        
class coche(vehiculo):
    '''Clase propia para los coches'''
    def contaminacion(self):
        print('Contamina mucho')
        
    def alquiler(self):
        print('Alquiler cuesta 100 e')
        
class tren(vehiculo):
    '''Clase propia para los trenes'''
    def contaminacion(self):
        print('Contamina menos')
        
    def comprar_billete(self):
        print('contaminacion')
```

## Operaciones Vectoriales

La implementación de Python que tenemos por defecto es **CPython**. Hay otras implementaciones de Python como **PyPy** o **Cython** que intentan superar estas dificultades, pero **CPython** es la más aceptada. 

```python
import platform 
platform.python_implementation()
```

Debido a la naturaleza dinámica de Python (no es necesario declarar las variables), este funciona más lentamente que C. 

NumPy tiene una serie de funciones para operar con arrays que mejoran bastante la eficiencia. Principalmente por dos razones: 

- Los códigos de dichas funciones están ya compilados, y declaran las variables antes de realizar las operaciones.

- Dichas funciones están preparadas para vectorizar las operaciones, es decir, actuar sobre todo el array en vez de ir elemento por elemento. 

```python
a = np.random.randint(10,20,1000)

# Aproach 1
def mi_suma(x):
    suma = 0
    for i in x:
        suma += i
        
    return suma

%timeit suma = mi_suma(a)
    
# Approach 2

%timeit suma = a.sum()
```

Las operaciones típicas con nparrays se vectorizan automáticamente, y utilizan la misma sintaxis que Python, aunque también tienen su propio nombre.

```python
%timeit np.arange(5)/np.linspace(10,20,5)
%timeit np.divide(np.arange(5),np.linspace(10,20,5))
```

También hay operaciones numéricas más complejas, que en el caso de existir en Python toman el mismo nombre.

```python
a = np.random.normal(0,10,1000)
%timeit abs(a)
%timeit np.abs(a)
```

```python
def crear_df(cols,ind):
    """Crear un DataFrame de manera facil"""
    valores = []
    
    for i in ind:
        fila = []
        for c in range(len(cols)):
            fila.append(str(cols[c])+str(i))
            
        valores.append(fila)
        
    valores = np.array(valores)        
    return pd.DataFrame(valores,columns = cols)
            
crear_df(['A','B','C'],range(3))
```

- **applymap**: aplica una funcion a cada elemento de un dataframe
- **transform**: aplica una transformacion capaz de mezclar elementos con agregaciones de datos

## Groupby

Un group by tiene en realidad tres fases: 
- Separar
- Aplicar un cálculo
- Combinar

Lo primero que hacemos es agrupar el dataframe (esto sería el *by*)

```python
planetas.groupby('method',as_index = False)
```

Una vez agrupado el dataframe calculamos cosas

```python
planetas.groupby('method',as_index = False).median()
planetas.groupby('method',as_index = False).max()
# gracias a que las clases heredan los metodos, el describe tambien se hace por grupos
planetas.groupby('method',as_index = False).describe().unstack().head()
```

La agregacion es un tipo de groupby en el que se reduce cada grupo a un escalar

```python
grouped = planetas.groupby('method',as_index = False).agg('sum')
grouped
```

Si queremos agrupar cada columna con una operación

```python
grouped = planetas.groupby('method',as_index=False).agg({'number':'sum', 
                                                         'orbital_period':['mean','median'],
                                                         'mass':'mean', 
                                                         'year':'max'})
grouped['orbital_period']['mean']
```

Las operaciones funcionales también se ven afectadas por el groupby

```python
planetas[['method','mass']].groupby('method', as_index = True).sum()
# Masa normalizada por grupos
planetas[['method','mass']].groupby('method', as_index = True).transform(lambda x: abs(x - x.mean())/x.std()).head()
```

Como habíamos comentado hay missings en los datos. 

```python
planetas.isna().sum()
```

Podemos imputar esos datos faltantes con una media por grupo. 

```python
planetas['mass'] = planetas[['method','mass']].groupby('method', as_index = True).transform(lambda x: x.fillna(x.mean()))
planetas.isna().sum()
```

```python
""" Multi-pendulum Harmonograph simulator using numpy and matplotlib
 
    You can specify any number of pendulums npend > 0; this number also sets
    the number of frequencies available. The sine wave parameters are 
    a: amplitude, a random float in the range 0 to 1;
    f: frequency, a random near-integer in the range 1 to npend
    p: phase, a random float in the range 0 to 2pi
 
    Copyright 2017 Alan Richmond @ Python3.codes
    The MIT License https://opensource.org/licenses/MIT
"""
import random as r
import matplotlib.pyplot as plt
from numpy import arange, sin, cos, exp, pi
plt.rcParams["figure.figsize"] = 8,6    # size of plot in inches
 
mf = npend = 4          # # of pendulums & maximum frequency
sigma = 0.005           # frequency spread (from integer)
step = 0.01             # step size
steps = 40000           # # of steps
linew = 2               # line width
def xprint(name, value):    # convenience function to print params.
    print(name+' '.join(['%.4f' % x for x in value]))
 
t = arange(steps)*step      # time axis
d = 1 - arange(steps)/steps # decay vector
while True:
    n = input("Number of pendulums (%d)(0=exit): "%npend)
    if n != '': npend = int(n)
    if npend == 0: break
    n = input("Deviation from integer freq.(%f): "%sigma)
    if n != '': sigma = float(n)
    ax = [r.uniform(0, 1) for i in range(npend)]
    ay = [r.uniform(0, 1) for i in range(npend)]
    px = [r.uniform(0, 2*pi) for i in range(npend)]
    py = [r.uniform(0, 2*pi) for i in range(npend)]
    fx = [r.randint(1, mf) + r.gauss(0, sigma) for i in range(npend)]
    fy = [r.randint(1, mf) + r.gauss(0, sigma) for i in range(npend)]
    xprint('ax = ', ax); xprint('fx = ', fx); xprint('px = ', px)
    xprint('ay = ', ay); xprint('fy = ', fy); xprint('py = ', py)
    x = y = 0
    for i in range(npend):
        x += d * (ax[i] * sin(t * fx[i] + px[i]))
        y += d * (ay[i] * sin(t * fy[i] + py[i]))
    plt.figure(facecolor = 'white')
    plt.plot(x, y, 'k', linewidth=1.5)
    plt.axis('off')
    plt.subplots_adjust(left=0.0, right=1.0, top=1.0, bottom=0.0)
    plt.show(block=False)
```

# scikit-learn

Antes de utilizar algún modelo de predicción, tenemos que tratar un poco la tabla. Para empezar nos quedamos con las variables que nos interesan. 

```python
titanic.isnull().sum()
```

```python
# titanic = titanic.drop(columns=['PassengerId' , 'Name' , 'Ticket' , 'Cabin'] )
titanic = titanic.drop(labels=['PassengerId' , 'Name' , 'Ticket' , 'Cabin'] , axis = 1)
titanic.head()
```

Los pasajeros que no tienen embarked se les imputa con la moda.

```python
titanic[titanic['Embarked'].isnull()]
```

```python
moda_embark = titanic['Embarked'].value_counts().idxmax()

titanic[['Embarked']] = titanic[['Embarked']].fillna(moda_embark) 
```

```python
titanic.isnull().sum()
```

Los pasajeros que no tienen edad se les imputa la media por clase y genero.

```python
media_edad = titanic[['Pclass','Sex','Age']].groupby(by=['Pclass','Sex'] , as_index=False).mean()
media_edad.columns = ['Pclass','Sex','Age_media']

titanic = pd.merge(titanic,media_edad,how = 'left')
titanic.head(10)
```

```python
titanic['Age'] = titanic.apply(lambda x: x.Age_media if np.isnan(x.Age) else x.Age,  axis  = 1 )
# titanic = titanic.drop(columns='Age_media')
titanic = titanic.drop(labels=['Age_media'] , axis = 1)
```

```python
titanic.isnull().sum()
```

Variables dummies para las string

```python
titanic = pd.get_dummies(data=titanic , columns=['Embarked' , 'Sex'])
```

```python
y = titanic.Survived

variables = list(set(titanic.columns)-set(['Survived']))
X = titanic[variables]
```

Partir en train y test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.15, 
                                                    random_state=4,
                                                    stratify = y # partición estratificada por la variable target
                                                   )
X_train.shape
```

Regresion logística

```python
from sklearn import linear_model
from sklearn.metrics import zero_one_loss

reg = linear_model.LogisticRegression(fit_intercept=True ).fit(X = X_train ,y = y_train)
reg.coef_
reg.intercept_

y_pred = reg.predict(X_test)
error_rate = zero_one_loss(y_test, y_pred) # fraccion de registros clasificados erroneamente
accuracy = 1 - error_rate

print(accuracy)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

n_estimators = 5000

GBC = GradientBoostingClassifier(learning_rate = 0.001, 
                                 n_estimators = n_estimators , 
                                 max_depth = 2,
                                 subsample = 0.85, 
                                 presort = True)\
                                 .fit(X = X_train, y = y_train)
    
y_pred = GBC.predict(X_test)
error_rate = zero_one_loss(y_test, y_pred) # fraccion de registros clasificados erroneamente
accuracy = 1 - error_rate

print(accuracy)
```
