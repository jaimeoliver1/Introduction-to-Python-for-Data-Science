```python
import pandas as pd
import numpy as np
import gzip
from io import StringIO,BytesIO
from collections import defaultdict
import csv
from datetime import datetime
import os,sys
```

```python
ejem1_col = ['Transaccion','Cliente','Fecha','Producto','Grupo','Venta','Pago','Pais']
ejem2_col = ['Cliente','Impuestos','Tarjeta','Origen']
file1 = 'prueba.csv'
file2 = 'prueba2.csv'
```

```python
data1 = pd.read_csv(file1,sep=',',error_bad_lines=False,low_memory=False,encoding='utf-8',names=ejem1_col)
data2 = pd.read_csv(file2,sep=',',error_bad_lines=False,low_memory=False,encoding='utf-8',names=ejem2_col)
```

```python
len(list(set(data1.Cliente.tolist()) ^ set(data2.Cliente.tolist())))
```

```python
tstart = datetime.now()
data = data1.merge(data2,how='inner',on='Cliente')
print('Time: ' + str(datetime.now() - tstart))
```

```python
dict_2 = data2.set_index('Cliente')[['Impuestos','Tarjeta','Origen']].to_dict('index')
dict_2
```

```python
tstart = datetime.now()
def calculoImpuestos(x):
    return x.Venta * (1. + dict_2[x.Cliente]['Impuestos']/100.)
data1['totalImpuestos'] = data1.apply(lambda x: calculoImpuestos(x), axis=1)
print('Time: ' + str(datetime.now() - tstart))
```

```python
tstart = datetime.now()
def calculoImpuestos(x):
    return x.Venta * (1. + dict_2[x.Cliente]['Impuestos']/100.)
data1['totalImpuestos1'] = data1[['Cliente','Venta']].apply(lambda x: calculoImpuestos(x), axis=1)
print('Time: ' + str(datetime.now() - tstart))
```

```python
tstart = datetime.now()
def calculoImpuestos(x):
    if x.Pais != 'USA':
        return x.Venta * (1. + dict_2[x.Cliente]['Impuestos']/100.)
    else:
        return x.Venta
data1['totalImpuestos2'] = data1.apply(lambda x: calculoImpuestos(x), axis=1)
print('Time: ' + str(datetime.now() - tstart))
```

```python
tstart = datetime.now()
def calculoImpuestos(x):
    if x.Pais != 'USA':
        return x.Venta * (1. + dict_2[x.Cliente]['Impuestos']/100.)
    else:
        return x.Venta
data1['totalImpuestos3'] = data1[['Cliente','Venta','Pais']].apply(lambda x: calculoImpuestos(x), axis=1)
print('Time: ' + str(datetime.now() - tstart))
```

```python
tstart = datetime.now()
data1['Tarjeta'] = data1['Cliente'].map(lambda x: dict_2[x]['Tarjeta'] if x in dict_2 else np.nan, na_action='ignore')
print('Time: ' + str(datetime.now() - tstart))
```
